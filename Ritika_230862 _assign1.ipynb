```python
# question 1
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 4 * np.pi, 400)

y1 = np.sin(x)
y2 = np.cos(x / 2)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x/2)', color='red', linestyle='--')

plt.title("Sin vs. Cos⁄2")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()

plt.savefig("sin_vs_cos2.png", dpi=300)

plt.show()
```

```python
#question 2
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

mean_sepal = df.groupby("species")["sepal_length"].mean().round(2)

plt.figure(figsize=(8, 5))
bars = plt.bar(mean_sepal.index, mean_sepal.values, color=["skyblue", "lavender", "lightgreen"])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.02, f'{height:.2f}', ha='center', fontsize=12)

plt.xlabel("Species")
plt.ylabel("Mean Sepal Length in (cm)")
plt.title("Mean Sepal Length by Species")
plt.ylim(0, max(mean_sepal.values) + 0.5)

plt.tight_layout()
plt.show()
```

```python
#question 3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/CarlHatoum/Spotify-TOP-50-songs/master/top50.csv"
df = pd.read_csv(url, encoding='latin1')

features = ['Danceability', 'Energy', 'Valence.']
df_subset = df[features + ['Genre']]

sns.set_style("darkgrid")

sns.pairplot(df_subset, hue='Genre', corner=True, palette='husl')

plt.suptitle("Danceability vs Energy vs Valence", y=1.02)
plt.tight_layout()
plt.show()
```

```python
#question 4
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
A = np.random.normal(0, 1, 1000)
B = np.random.normal(2, 1.5, 1000)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.hist(A, bins=30, alpha=0.6, label="A ~ N(0, 1)", color='skyblue', edgecolor='black')
ax1.hist(B, bins=30, alpha=0.6, label="B ~ N(2, 1.5)", color='pink', edgecolor='black')
ax1.set_title("Histograms")
ax1.set_xlabel("Value")
ax1.set_ylabel("Frequency")
ax1.legend()

ax2.boxplot([A, B], tick_labels=["A", "B"])
ax2.set_title("Boxplot Comparison")
ax2.set_ylabel("Value")

plt.suptitle("Comparing Two Normal Distributions", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
#question 5
# Important!!  this code is compatible with google colab and pycharm, for vscode some functions need to be changed,
# so i am including an HTML doc that contains the animation of this program
# this is the link of google colab notebook and can be checked
# https://colab.research.google.com/drive/1xf3TgyFNOuH899IHI-g3V0b6YGdQM3BI?usp=sharing
import plotly.express as px
import pandas as pd

df = px.data.gapminder()

df_2007 = df[df["year"] == 2007]

continent_avg = df_2007.groupby("continent")[["lifeExp", "gdpPercap"]].mean().reset_index()
print("Average by continent:\n", continent_avg)

fig_static = px.scatter(
    df_2007,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    size_max=60,
    log_x=True,
    title="Gapminder 2007: GDP vs Average Life Expectancy"
)

fig_static.show()

fig_animated = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    animation_frame="year",
    animation_group="country",
    size="pop",
    color="continent",
    hover_name="country",
    size_max=60,
    log_x=True,
    range_x=[100, 100000],
    range_y=[20, 90],
    title="Gapminder: GDP per Capita vs Life Expectancy (1960–2007)"
)

fig_animated.show()
```
